```python
import pandas as pd
df1 = pd.read_csv('EST_Tae_CCR_Rawdata_Figure1.csv')
print(df1.shape)
print(df1.head())
```

```python
# Based on README: columns X-AF are the clean averaged data
# In pandas these are roughly columns 23 onward
# Let's isolate and look at them

df_clean = df1.iloc[:, 23:]
print(df_clean.head(10))
```

```python
# Unnamed: 23 appears to be 50C normalized values
# Unnamed: 24 appears to be 50C std dev
# Unnamed: 26 appears to be 30C normalized
# Unnamed: 29 appears to be 2C normalized

df_50c = df1[['Unnamed: 23', 'Unnamed: 24']].dropna()
df_50c.columns = ['concentration', 'std_dev']
print(df_50c)
```

```python
# Add a replicate label based on where 0 appears in time column
df_50c = df1[['Unnamed: 23', 'Unnamed: 24']].copy()
df_50c.columns = ['time_days', 'concentration']
df_50c['time_days'] = pd.to_numeric(df_50c['time_days'], errors='coerce')
df_50c['concentration'] = pd.to_numeric(df_50c['concentration'], errors='coerce')
df_50c = df_50c.dropna()

# Assign replicate numbers
df_50c['replicate'] = (df_50c['time_days'] == 0).cumsum()
df_50c['temperature'] = '50C'
print(df_50c)
```
